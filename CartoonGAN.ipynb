```python
import matplotlib.pyplot as plt
from PIL import Image

# Load the original and output images
original_img = Image.open('/content/cartoon-gan/resized_image.jpg')  # Your original image file name
output_img = Image.open('./output_folder/output_image.jpg')  # The output avatar image

# Create a figure to display both images
fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns

# Display the original image
axes[0].imshow(original_img)
axes[0].set_title('Original Image')
axes[0].axis('off')  # Hide axes

# Display the output image (avatar)
axes[1].imshow(output_img)
axes[1].set_title('Avatar Image')
axes[1].axis('off')  # Hide axes

# Show the plot
plt.show()
```
